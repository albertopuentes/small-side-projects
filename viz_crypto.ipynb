```python
import pandas_datareader as web
import matplotlib.pyplot as plt
import mplfinance as mpf
import datetime as dt
```

```python
# Crypto Pair
crypto = 'BTC'
currency = 'USD'
```

```python
# Scrape data from Yahoo Finance
start= dt.datetime(2020,1,1)
end= dt.datetime.now()

data = web.DataReader(f'{crypto}-{currency}', 'yahoo', start, end)
```

```python
print(data)
```

```python
# plot data
plt.plot(data['Close'])
plt.show()
```

```python
# candle chart
mpf.plot(data[-25:], type='candle', style='yahoo', volume=True)
```

```python
#chart multiple crypto
btc = web.DataReader(f'{"BTC"}-{currency}', 'yahoo', start, end)
eth = web.DataReader(f'{"ETH"}-{currency}', 'yahoo', start, end)
```

```python
# visualize in log scale
plt.plot(btc['Close'], label='BTC')
plt.plot(eth['Close'], label='ETH')
plt.yscale('log')
plt.legend(loc='upper left')
```
